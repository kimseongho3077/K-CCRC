# 데이터 불러오기 

```python
import pandas as pd
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings(action='ignore')
```

```python
# 전처리 해두었던 시설별_전체 데이터 불러옴(시설 + 해당 시설이 있는 시군구의 인프라 + 해당 시설의 세대수)
df = pd.read_csv('../전처리된데이터/시설별_전체.csv')
df
```

```python
# 인덱스를 시설 이름으로 변경
df.set_index('이름', inplace = True)
df
```

```python
# 시설별_전체 데이터 확인
df.info()
```

# 다중 선형 회귀(sklearn 사용, scaling X) 

```python
# 모델 학습을 위해 train과 test를 나눌 때 사용
from sklearn.model_selection import train_test_split

# sklearn의 Linear_Regression 클래스 이용
from sklearn.linear_model import LinearRegression

# 평가지표로는 MSE(RMSE), R2 이용
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# 법정동 제거 후 , X(독립변수), y(종속변수) 나눠줌
copy_df = df.copy()
copy_df.drop('법정동코드', axis = 1, inplace = True)
X_data_all = copy_df.drop('세대수', axis = 1)
y_target_all = copy_df['세대수']
```

```python
# 기본 다중선형회귀 (LinearRegression)
X_train_all, X_test_all, y_train_all, y_test_all = train_test_split(X_data_all, y_target_all, test_size = 0.3, random_state = 156)

lr1 = LinearRegression()
lr1.fit(X_train_all, y_train_all)

y_pred_all = lr1.predict(X_test_all)
mse = mean_squared_error(y_test_all, y_pred_all)
rmse = np.sqrt(mse)

print('MSE : {0:.3f}, RMSE : {1:.3f}'.format(mse, rmse))
print('R2 score : {0:.3f}'.format(r2_score(y_test_all, y_pred_all)))
```

# 다중 선형 회귀(sklearn 사용, standard scaling) 

```python
# 스케일러 라이브러리 import
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
# standardscaler 적용
scaler_std = StandardScaler()
scaler_std.fit(copy_df)
copy_std = scaler_std.transform(copy_df)

df_std = pd.DataFrame(copy_std, columns=copy_df.columns, index=list(copy_df.index.values))
df_std
```

```python
# 타겟값인 세대수만 뽑아서 X와 y 분리
X_std = df_std.drop('세대수', axis = 1)
y_std = df_std['세대수']
```

```python
# 스케일러 적용된 것으로 다중선형회귀(linearregression)
X_train_std, X_test_std, y_train_std, y_test_std= train_test_split(X_std, y_std, test_size = 0.3, random_state = 156)

lr_std = LinearRegression()
lr_std.fit(X_train_std, y_train_std)

y_pred_std = lr_std.predict(X_test_std)
mse = mean_squared_error(y_test_std, y_pred_std)
rmse = np.sqrt(mse)

print('MSE : {0:.3f}, RMSE : {1:.3f}'.format(mse, rmse))
print('R2 score : {0:.3f}'.format(r2_score(y_test_std, y_pred_std)))
```

# K-CCRC 입지 선정을 위해 수도권, 대도시는 제거

```python
총인구 = pd.read_csv('../전처리된데이터/시군구별_수요.csv')
총인구
```

```python
# 특별시, 광역시, 경기도 (대도시, 수도권) 이 들어가면 제거
총인구 = 총인구[~총인구['시도명'].str.contains("특별시", na=False, case=False)]
총인구 = 총인구[~총인구['시도명'].str.contains("광역시", na=False, case=False)]
총인구 = 총인구[~총인구['시도명'].str.contains("경기도", na=False, case=False)]
총인구
```

```python
# 인구 수 50만 이상의 경우 대도시
대도시제외 = 총인구[총인구['총_인구수'] <= 500000]
대도시제외
```

```python
후보지 = 세대수_결과.copy()
후보지
```

```python
# 후보지 결과와 대도시 제외한 지역 merge (대도시 수도권이 아닌 후보지)
최종후보지 = pd.merge(후보지, 대도시제외, on = '법정동코드', how = 'inner')
최종후보지
```

```python
# 필요한 columns만 추출
최종후보지 = 최종후보지[['법정동코드', '지역명_x', '세대수(log)', '세대수', '총_인구수']]
최종후보지
```

```python
# column명 정리
최종후보지.rename(columns = {'지역명_x' : '지역명'}, inplace = True)
최종후보지
```

```python
# 법정동코드를 인덱스로 설정
최종후보지.set_index('법정동코드', inplace = True)
```
