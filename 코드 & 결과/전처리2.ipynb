## 시군구별 대학교수 

```python
# 대학교 데이터 불러오기
대학교 = pd.read_excel('../원본데이터/교육시설/학교개황(20220310기준).xls')
대학교
```

```python
대학교.columns
```

```python
# 학교가 폐교하였거나 대학원인 행 제거
대학교 = 대학교[~대학교['학교상태'].str.contains("폐교", na=False, case=False)]
대학교 = 대학교[~대학교['학교명'].str.contains("대학원", na=False, case=False)]
```

```python
학교명 = 대학교.loc[:,'학교명']
주소 = 대학교.loc[:,'주소']
```

```python
# 학교명과 주소 concat
대학교 = pd.concat([학교명, 주소], axis = 1)
대학교
```

```python
대학교.isnull().sum()
```

```python
# 주소를 띄어쓰기 기준으로 시도명 / 시군구명으로 분리
분리 = 대학교['주소'].str.split(' ')
시도명 = 분리.str.get(0)
시군구명 = 분리.str.get(1)

대학교['시도명'], 대학교['시군구명'] = 시도명, 시군구명
대학교
```

```python
# 지역명을 '시도명_시군구명' 형태로 설정
대학교['지역명'] = 대학교['시도명'] + '_' + 대학교['시군구명']
대학교
```

```python
# 시도명이 '세종특별자치시'인 경우, 시군구명도 '세종특별자치시'로 변경
대학교.loc[대학교.시도명 == '세종특별자치시', '시군구명'] = '세종특별자치시'
대학교.loc[대학교.시도명 == '세종특별자치시', '지역명'] = '세종특별자치시_세종특별자치시'
대학교[대학교.시도명 == '세종특별자치시']
```

```python
# null 값 확인
대학교.isnull().sum()
```

```python
# 지역명 기준으로 대학교 수 count
대학교_수 = 대학교.groupby(대학교['지역명'], as_index=False)['학교명'].count()
대학교_수.drop(index = [0], inplace=True)
대학교_수
```

```python
# 유일지역 확인
유일지역_idx = 대학교_수['지역명'].value_counts() == 1
유일지역 = 유일지역_idx[유일지역_idx]
유일지역
```

```python
# 유일지역 데이터 전체 수정

대학교.loc[대학교.지역명 == '충남_아산시', '시도명'] = '충청남도'
대학교.loc[대학교.지역명 == '충남_아산시', '시군구명'] = '아산시'
대학교.loc[대학교.지역명 == '충남_아산시', '지역명'] = '충청남도_아산시'

대학교.loc[대학교.지역명 == '경남_진주시', '시도명'] = '경상남도'
대학교.loc[대학교.지역명 == '경남_진주시', '시군구명'] = '진주시'
대학교.loc[대학교.지역명 == '경남_진주시', '지역명'] = '경상남도_진주시'

대학교.loc[대학교.지역명 == '부산_금정구', '시도명'] = '부산광역시'
대학교.loc[대학교.지역명 == '부산_금정구', '시군구명'] = '금정구'
대학교.loc[대학교.지역명 == '부산_금정구', '지역명'] = '부산광역시_금정구'

대학교.loc[대학교.지역명 == '인천광역시_남구', '시도명'] = '인천광역시'
대학교.loc[대학교.지역명 == '인천광역시_남구', '시군구명'] = '미추홀구'
대학교.loc[대학교.지역명 == '인천광역시_남구', '지역명'] = '인천광역시_미추홀구'
```

```python
# 지역명 기준으로 대학교 수 count
대학교_수 = 대학교.groupby(대학교['지역명'], as_index=False)['학교명'].count()
대학교_수.drop(index = [0], inplace=True)
대학교_수.rename(columns = {'학교명' : '대학교_수'}, inplace = True)
대학교_수
```

```python
# 법정동코드와 대학교 수 merge
시군구별_대학교수 = pd.merge(법정동코드, 대학교_수, on = '지역명', how = 'outer')
시군구별_대학교수 = 시군구별_대학교수.fillna(0)
시군구별_대학교수['대학교_수'] = 시군구별_대학교수['대학교_수'].astype('int64')
시군구별_대학교수
```

```python
# 시군구별_초중고수와 시군구별_대학교수를 통해 시군구별_교육시설 인프라 생성
시군구별_교육시설 = pd.concat([시군구별_초중고수, 시군구별_대학교수['대학교_수']], axis = 1)
시군구별_교육시설
```

```python
# 법정동코드를 index로 설정
시군구별_교육시설.set_index('법정동코드', inplace = True)
```

```python
# 시군구별_교육시설 파일 반환
시군구별_교육시설.to_csv('../전처리된데이터/시군구별_교육시설.csv')
```

# 시군구별 교통시설

```python
# 대중교통 서비스별 만족도 현황 데이터 불러오기
교통 = pd.read_csv('../원본데이터/교통시설/한국교통안전공단_전국 지역별 대중교통 서비스별 만족도 현황_20201231.csv', encoding='cp949')
```

```python
교통
```

```python
# 구분1과 구분2를 합쳐서 구분으로 저장
교통['구분'] = 교통['구분1'] + '_' + 교통['구분2']
교통
```

```python
# 구분을 index로 설정
교통.set_index('구분', inplace = True)
교통
```

```python
# 구분1과 구분2 drop
교통.drop(['구분1', '구분2'], axis = 1, inplace = True)
교통
```

```python
# index 설정 해제
교통.reset_index(inplace = True)
교통
```

```python
# 안전환경서비스를 포함하는 행 제거 (결측치 다수)
교통 = 교통[~교통['구분'].str.contains("안전환경서비스", na=False, case=False)]
교통
```

```python
# 구분을 index로 설정
교통.set_index('구분', inplace = True)
교통
```

```python
교통 = 교통.T
```

```python
교통
```

```python
# 교통 데이터에 MinMaxScaler 적용
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(교통)
교통_정규화 = scaler.transform(교통)

# transform( )시 scale 변환된 데이터 셋이 numpy ndarry로 반환되어 이를 DataFrame으로 변환
교통시설 = pd.DataFrame(data=교통_정규화, columns = 교통.columns, index = 교통.index)
```

```python
교통시설
```

```python
# 각 만족도의 평균치 산출
시도별_교통 = 교통.mean(axis = 1)
시도별_교통 = pd.DataFrame(시도별_교통, columns = ['대중교통_만족도'])
시도별_교통
```

```python
시도별_교통.index
```

```python
시도별_교통
```

```python
# 각 데이터의 index를 법정동코드의 시도명으로 설정
시도별_교통.index = 법정동코드['시도명'].unique()
```

```python
시도별_교통
```

```python
# index 설정 해제 및 column 이름 변경
시도별_교통.reset_index(inplace = True)
시도별_교통.columns = ['시도명', '대중교통_만족도']
시도별_교통
```

```python
# 법정동코드와 시도별_교통 merge
시군구별_교통시설 = pd.merge(법정동코드, 시도별_교통, on = '시도명', how = 'inner')
시군구별_교통시설
```

```python
# 법정동코드로 index 설정
시군구별_교통시설.set_index('법정동코드', inplace = True)
```

```python
# 시군구별_교통시설 파일로 반환
시군구별_교통시설.to_csv('../전처리된데이터/시군구별_교통시설.csv')
```
