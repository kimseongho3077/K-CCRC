# 시군구별_법정동코드

```python
# 법정동 코드 불러오기
법정동코드 = pd.read_excel('../원본데이터/법정동코드/행정동_20220701.xlsx')
법정동코드.rename(columns = {'행정동코드' : '법정동코드'}, inplace = True)
법정동코드
```

```python
# 읍면동 단위 제외하고 시군구 단위만 추출. ~출장소는 제외
시군구별_법정동코드 = 법정동코드[(법정동코드['시군구명'].notnull()) & (법정동코드['읍면동명'].isnull())]
시군구별_법정동코드 = 시군구별_법정동코드[~시군구별_법정동코드['시군구명'].str.contains("출장소", na=False, case=False)]
시군구별_법정동코드.drop(['읍면동명', '생성일자', '말소일자'], axis = 1, inplace = True)
시군구별_법정동코드
```

```python
# 시군구명 분리 ex) '포항시 남구' -> '포항시', '남구'
분리 = 시군구별_법정동코드['시군구명'].str.split(' ')
시군구명1 = 분리.str.get(0)
시군구명2 = 분리.str.get(1)

시군구별_법정동코드['시군구명1'], 시군구별_법정동코드['시군구명2'] = 시군구명1, 시군구명2
시군구별_법정동코드
```

```python
# 시군구 단위의 변환을 위해 시군구명2가 null인 행은 제거한 후 '지역명' column에 '시도명_시군구명' 형태로 저장
시군구별_법정동코드 = 시군구별_법정동코드[시군구별_법정동코드['시군구명2'].isnull()]
시군구별_법정동코드.drop(['시군구명1', '시군구명2'], axis = 1, inplace = True)
시군구별_법정동코드['지역명'] = 시군구별_법정동코드['시도명'] + '_' + 시군구별_법정동코드['시군구명']
시군구별_법정동코드
```

```python
# 법정동코드를 index로 설정
시군구별_법정동코드.set_index('법정동코드', inplace = True)
```

```python
# 시군구별_법정동코드 파일 반환
시군구별_법정동코드.to_csv('../전처리된데이터/시군구별_법정동코드.csv')
```

# 시군구별_수요 

```python
# 연령별 인구 현황 파일 불러오기
인구 = pd.read_excel('../원본데이터/수요/202207_202207_연령별인구현황_월간.xlsx', header = 3, thousands = ',')
인구
```

```python
인구.columns
```

```python
인구.info()
```

```python
# 인구 수 '계' 까지만 추출
인구 = 인구.iloc[:, :25]
인구
```

```python
# 나중에 drop할 인구 수 '60세 미만' 추출
인구_60세미만 = 인구.iloc[:, 4:16]
인구_60세미만
```

```python
# 인구 수 '60세 미만' drop
인구.drop(인구_60세미만, axis = 1, inplace = True)
인구
```

```python
# 필요없는 '연령구간인구수' drop
인구.drop(['연령구간인구수'], axis = 1, inplace = True)
인구
```

```python
# 행정기관을 띄어쓰기 기준으로 시도 / 시군구 / 동명으로 분리
분리 = 인구.행정기관.str.split(' ')
시도 = 분리.str.get(0)
시군구 = 분리.str.get(1)
동명 = 분리.str.get(2)

인구['시도명'], 인구['시군구명'], 인구['동명'] = 시도, 시군구, 동명
```

```python
# 행정기관코드 str형으로 변경
인구.iloc[:, 0] = 인구.iloc[:, 0].astype('str')
인구.rename(columns = {'행정기관코드' : '법정동코드'}, inplace = True)
인구
```

```python
# '60~64세'부터 '100세 이상'까지 더해서 '60세 이상' 피쳐 생성
인구['60세이상'] = np.sum(인구.iloc[:, 3:12], axis = 1)
```

```python
# 필요한 피쳐만 가져오기
인구 = 인구[['법정동코드','총 인구수', '시도명', '시군구명', '동명', '60세이상']]
인구
```

```python
# 동명, 시군구명에서 빈값을 NaN으로 변경
인구.동명.replace('',np.nan, regex = True, inplace = True)
인구.시군구명.replace('',np.nan, regex = True, inplace = True)
인구
```

```python
#'세종특별자치시'의 세종시를 '세종특별자치시'로 설정 (법정동코드 활용)
인구.loc[(인구.시도명 == '세종특별자치시') & (인구.법정동코드 == '3611000000'), '시군구명'] = '세종특별자치시'
인구[인구.시도명 == '세종특별자치시']
```

```python
# '동명'은 NaN값이면서 '시군구명'은 NaN값이 아닌, 시군구 단위만 뽑아냄
# 어차피 NaN값인 '동명'은 제거
인구 = 인구[인구['동명'].isnull()]
인구 = 인구[인구['시군구명'].isnull() == False]
인구.drop('동명', axis = 1, inplace = True)
인구
```

```python
# 시군구명이 ~출장소인곳은 제거 예정
# 인천 3개, 경기도 5개, 전북 2개, 경남 3개
인구 = 인구[~인구['시군구명'].str.contains("출장소", na=False, case=False)]
인구
```

```python
# column 이름 재설정 후 총 인구 수와 60세 이상 인구 수를 통해'60세이상_비율' 산출 후 column 추가
인구.rename(columns = {'60세이상' : '60세이상_인구수', '총 인구수' : '총_인구수'}, inplace = True)
인구['60세이상_비율'] = 인구['60세이상_인구수'] / 인구['총_인구수']
인구
```

```python
# 지역명을 '시도명_시군구명' 형태로 변경
인구['지역명'] = 인구['시도명'] + '_' + 인구['시군구명']
시군구별_수요 = 인구[['법정동코드', '시도명', '시군구명', '지역명', '총_인구수', '60세이상_인구수', '60세이상_비율']]
시군구별_수요
```

```python
# 법정동코드로index 설정
시군구별_수요.set_index('법정동코드', inplace = True)
```

```python
# 시군구별_수요 파일 반환
시군구별_수요.to_csv('../전처리된데이터/시군구별_수요.csv')
```

# 시군구별_공급

```python
# 시군구별 공급 정도의 확인을 위해 시군구별 노인복지주택 수 추출
공급_노인복지주택 = pd.read_excel('../원본데이터/공급/2022_목차_이용안내_총괄표(시.군.구)_.xlsx', header = 4, sheet_name = 0, index_col = 0)
공급_노인복지주택
```

```python
공급_노인복지주택.info()
```

```python
공급_노인복지주택.reset_index(inplace = True)
공급_노인복지주택
```

```python
# 원본데이터에서 세종시 -> 세종특별자치시로 변경
공급_노인복지주택.loc[공급_노인복지주택['index'] == '세종시', 'index'] = '세종특별자치시'
공급_노인복지주택[공급_노인복지주택['index'] == '세종특별자치시']
```

```python
# 현 존재하는 모든 시도명 값을 가진 리스트 생성
시도명 = ['서울특별시', '부산광역시', '대구광역시', '인천광역시', '광주광역시', '대전광역시', '울산광역시', '세종특별자치시', '경기도', '강원도', '충청북도', '충청남도', '전라북도', '전라남도', '경상북도', '경상남도', '제주특별자치도']
len(시도명)
```

```python
# '합계'가 들어간 데이터 인덱스 값을 가진 리스트 생성
시도_idx = 공급_노인복지주택[공급_노인복지주택['index'].str.contains('합계')].index
시도_idx
```

```python
# 시도명 값을 저장할 리스트 생성

l = []

# 서울특별시 ~ 경상남도
for i in range (0, len(시도_idx)-1): # 0 ~ 15
    a = 시도_idx[i] # 0번째 14번째
    b = 시도_idx[i+1] # 1번째 15번째
    for j in range(a, b):
        l.append(시도명[i])

# 제주특별자치도
for k in range(시도_idx[16], len(공급_노인복지주택)):
    l.append(시도명[16])
    
l
```

```python
지역명 = pd.DataFrame(l)
지역명.columns = ['지역명']
지역명
```

```python
# 새로 만든 지역명 df과, 원본 데이터의 index(시군구)값과, 정원.2(노인복지주택정원)값만 추출하여 concat. 새로운 노인복지주택_정원이라는 df생성
노인복지주택_정원 = pd.concat([지역명, 공급_노인복지주택[['index', '정원.2']]], axis = 1)
노인복지주택_정원
```

```python
# column 이름 정리
노인복지주택_정원 = 노인복지주택_정원.rename(columns = {'지역명' : '시도명', 'index' : '시군구명', '정원.2' : '노인복지주택_정원'})
노인복지주택_정원
```

```python
# 시군구 data가 ~합계인 값들 모두 제거
노인복지주택_정원 = 노인복지주택_정원[~노인복지주택_정원['시군구명'].str.contains("합계", na=False, case=False)]
노인복지주택_정원
```

```python
# 시도와 시군구값을 문자열 합치기 해주어서 새로운 '지역명' 피쳐 생성
노인복지주택_정원['지역명'] = 노인복지주택_정원['시도명'] + '_' + 노인복지주택_정원['시군구명']
노인복지주택_정원 = 노인복지주택_정원[['지역명', '노인복지주택_정원']]
노인복지주택_정원
```

```python
# 법정동코드 파일 불러오기
법정동코드 = pd.read_csv('../전처리된데이터/시군구별_법정동코드.csv')
```

```python
# 시군구별 공급을 법정동코드와 병합
시군구별_공급 = pd.merge(법정동코드, 노인복지주택_정원, on = '지역명', how = 'inner')
시군구별_공급
```

```python
# 법정동코드로 index 설정
시군구별_공급.set_index('법정동코드', inplace = True)
```

```python
# 시군구별_공급 파일 반환
시군구별_공급.to_csv('../전처리된데이터/시군구별_공급.csv')
```

## 시군구별_우체국 

```python
# 시군구별 우체국 수 확인을 위해 파일 불러오기
우체국 = pd.read_csv('../원본데이터/공공시설/과학기술정보통신부 우정사업본부_전국 우체국 현황_20210630.csv', encoding='cp949')
우체국
```

```python
#우체국 = pd.DataFrame(우체국)
우체국 = 우체국.iloc[:,2:4]
우체국
```

```python
# '주소(지번)' column을 활용하여 각 우체국의 지역명 추출
우체국[['시도명','시군구명','그 외']] = pd.DataFrame(우체국[' 주소(지번) '].str.split(' ', 2).tolist())
우체국 = 우체국.iloc[:,0:4]
우체국.drop([' 주소(지번) '],axis=1,inplace=True)
우체국['지역명']= 우체국['시도명']+'_'+우체국['시군구명']
우체국
```

```python
# 지역명의 값이 1개인 행을 유일지역으로 저장 (통일되지 않은 데이터 값으로 인해 다른 지역으로 인식되는 경우)
유일지역_idx = 우체국['지역명'].value_counts() == 1
유일지역 = 유일지역_idx[유일지역_idx]
유일지역
```

### 시군구별 문화시설 최종

```python
# 시군구별 도서관, 문화시설 merge 하여 시군구별 문화시설 인프라 df 생성
시군구별_문화시설 = pd.merge(시군구별_도서관, 문화시설, on = '법정동코드', how = 'inner')
시군구별_문화시설
```

```python
# 시군구별 문화시설에 지역명도 함께 부여하기 위하여 merge
시군구별_문화시설 = pd.merge(법정동코드, 시군구별_문화시설, on = '법정동코드', how = 'inner')
시군구별_문화시설.set_index('법정동코드', inplace = True)
시군구별_문화시설
```
